```python
def fetch_cast(obj):
 lst = []
 counter = 0
 
 for i in ast.literal_eval(obj):
 # print(i['name'])
     if counter != 3:
         lst.append(i['name'])
         counter += 1
     else:
         break
 
 return lst
```

# NLP

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
 
 lst = []
 
 for i in text.split():
     lst.append(ps.stem(i))
 
 return " ".join(lst)
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
def get_recommended_movie(movie_name):
 try: 
     movie_index = new_df[new_df['title'] == movie_name].index[0]
     distance = similarity[movie_index]
     movie_list = sorted(list(enumerate(distance)), reverse=True, key= lambda x: x[1])
     if len(movie_list) >= 6:
        movie_list = movie_list[0:6]

     for i in movie_list:
    # print(i[0])
      print(new_df.iloc[i[0]].title) 
 except IndexError:
    print("Sorry! We do not have information regarding that movie.")
```
